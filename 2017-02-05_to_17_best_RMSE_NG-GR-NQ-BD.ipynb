```python
# read in daily difference datasets
# read in daily difference datasets
dif_06 = xr.open_dataarray('./dif_06_best.nc')
dif_07 = xr.open_dataarray('./dif_07_best.nc')
dif_08 = xr.open_dataarray('./dif_08_best.nc')
dif_09 = xr.open_dataarray('./dif_09_best.nc')
dif_10 = xr.open_dataarray('./dif_10_best.nc')
dif_11 = xr.open_dataarray('./dif_11_best.nc')
dif_12 = xr.open_dataarray('./dif_12_best.nc')
dif_13 = xr.open_dataarray('./dif_13_best.nc')
dif_14 = xr.open_dataarray('./dif_14_best.nc')
dif_15 = xr.open_dataarray('./dif_15_best.nc')

#dif_arr = [dif_05, dif_06, dif_07, dif_08, dif_09, dif_10, dif_11, dif_12, dif_13, dif_14, dif_15, dif_16, dif_17]
dif_arr = [dif_06, dif_07, dif_08, dif_09, dif_10, dif_11, dif_12, dif_13, dif_14, dif_15]
```

```python
# rmse
def rmse():
    sq_arr = []
    for i in dif_arr:
        sq_dif = numpy.square(i)
        sq_arr.append(sq_dif)
    sum_arr = sq_arr[0]
    for j in sq_arr[1:]:
        sum_arr += j
    mean_sq = sum_arr / 10
    rmse = numpy.sqrt(mean_sq)
    return rmse

rm = rmse()
rm
```

```python
# for reference
numpy.unique(rm), max(numpy.unique(rm)), min(numpy.unique(rm))
```

```python
# convert array to masked array to overlook NaNs
import numpy.ma as ma
m_out = ma.masked_invalid(rm)
m_out
```

```python
# rmse over entire region
def overall_rmse():
    sq_arr = []
    for i in dif_arr:
        sq_dif = numpy.square(i.data)
        sq_arr.append(sq_dif)
    arr_sq = [x for x in numpy.array(sq_arr).flatten() if str(x) != 'nan']
    div = sum(arr_sq) / (len(arr_sq))
    return numpy.sqrt(div)
overall_rmse()
```
